```python
import requests
from bs4 import BeautifulSoup

def extract_text_from_url(url):
    try:
        response = requests.get(url)
        
        # Parsing HTML Content
        soup = BeautifulSoup(response.content, 'lxml')
        
        # Extracting text from the HTML
        text = soup.get_text()
        
        return text
    except Exception as e:
        print(f"Error fetching URL {url}: {str(e)}")
        return None
```

```python
import os

output_dir = 'text_data'

os.makedirs(output_dir, exist_ok=True)

# Iterating through each row in DataFrame
for index, row in df.iterrows():
    url_id = row['URL_ID']
    url = row['URL']
    
    # Extracting text from URL
    text = extract_text_from_url(url)
    
    if text:
        filename = os.path.join(output_dir, f"{url_id}.txt")
        
        with open(filename, 'w', encoding='utf-8') as f:
            f.write(text)
            
        print(f"Text extracted from URL {url} saved to {filename}")
```

```python
import os

stop_words = set()

stop_words_dir = '/Users/abhi/Downloads/Assignment_DataScience/StopWords'

# Iterating over each file in the StopWords directory
for filename in os.listdir(stop_words_dir):
    if filename.endswith('.txt'):
        file_path = os.path.join(stop_words_dir, filename)
        with open(file_path, 'r', encoding='ISO-8859-1') as file:
            # Read stop words and add them to the set
            for line in file:
                stop_words.update(line.strip().split())

print(f"Loaded {len(stop_words)} stop words.")
```

```python
input_dir = 'text_data'
output_dir = 'cleaned_text_data'

# Creating the output directory
os.makedirs(output_dir, exist_ok=True)

# Iterating through each file in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith('.txt'):
        file_path = os.path.join(input_dir, filename)
        output_file_path = os.path.join(output_dir, filename)
        
        with open(file_path, 'r', encoding='utf-8') as input_file:
            text = input_file.read()
            
            # Removing stop words
            cleaned_text = ' '.join(word for word in text.split() if word.lower() not in stop_words)
            
            # Writing cleaned text to output file
            with open(output_file_path, 'w', encoding='utf-8') as output_file:
                output_file.write(cleaned_text)
                
        print(f"Cleaned text saved to {output_file_path}")
```

```python
import pandas as pd

nltk.download('punkt')

input_excel_path = '/Users/abhi/Downloads/Assignment_DataScience/Input.xlsx'
input_dir = 'cleaned_text_data'
output_excel_path = '/Users/abhi/Downloads/Assignment_DataScience/Input.xlsx'

df = pd.read_excel(input_excel_path)

# To Ensure the column names are correct
url_id_col = 'URL_ID' 

# Initialize lists for metrics
positive_scores = []
negative_scores = []
polarity_scores = []
subjectivity_scores = []
average_sentence_lengths = []
percentage_complex_words_list = []
fog_indices = []
average_words_per_sentence_list = []
complex_word_counts = []
word_counts = []
syllables_per_word_list = []
personal_pronouns_counts = []
average_word_lengths = []


def calculate_scores(text, positive_words, negative_words):
    tokens = nltk.word_tokenize(text)
    positive_score = sum(1 for token in tokens if token.lower() in positive_words)
    negative_score = sum(1 for token in tokens if token.lower() in negative_words)
    return positive_score, negative_score

def calculate_polarity(positive_score, negative_score):
    if positive_score + negative_score == 0:
        return 0  # To avoid division by zero
    polarity = (positive_score - negative_score) / ((positive_score + negative_score) + 0.000001)
    return polarity

def calculate_subjectivity(positive_score, negative_score, total_word_count):
    if positive_score + negative_score == 0:
        return 0  # To avoid division by zero
    subjectivity = (positive_score + negative_score) / (total_word_count + 0.000001)
    return subjectivity

def calculate_average_sentence_length(text):
    sentences = sent_tokenize(text)
    total_sentences = len(sentences)
    words = word_tokenize(text)
    total_words = len(words)
    return total_words / total_sentences if total_sentences > 0 else 0

def calculate_percentage_complex_words(text):
    translator = str.maketrans('', '', string.punctuation)
    cleaned_text = text.translate(translator)
    words = cleaned_text.split()
    total_words = len(words)
    complex_word_count = sum(1 for word in words if calculate_syllable_count(word) > 2)
    return (complex_word_count / total_words) * 100 if total_words > 0 else 0

def calculate_gunning_fog_index(average_sentence_length, percentage_complex_words):
    return 0.4 * (average_sentence_length + percentage_complex_words)

def calculate_syllable_count(word):
    vowels = "aeiouy"
    count = 0
    end_exceptions = ["es", "ed"]
    if any(word.endswith(end) for end in end_exceptions):
        return 0
    previous_char_was_vowel = False
    for char in word.lower():
        if char in vowels:
            if not previous_char_was_vowel:
                count += 1
                previous_char_was_vowel = True
        else:
            previous_char_was_vowel = False
    if word.endswith('e') and not word.endswith('le'):
        count -= 1
    return max(count, 1)

def count_syllables_in_text(text):
    translator = str.maketrans('', '', string.punctuation)
    cleaned_text = text.translate(translator)
    words = cleaned_text.split()
    return sum(calculate_syllable_count(word) for word in words)

def count_complex_words(text):
    translator = str.maketrans('', '', string.punctuation)
    cleaned_text = text.translate(translator)
    words = cleaned_text.split()
    return sum(1 for word in words if calculate_syllable_count(word) > 2)

def count_words_without_punctuation(text):
    translator = str.maketrans('', '', string.punctuation)
    cleaned_text = text.translate(translator)
    words = cleaned_text.split()
    return len(words)

def count_personal_pronouns(text):
    personal_pronouns_pattern = r'\b(?:I|we|my|ours|us)\b'
    matches = re.findall(personal_pronouns_pattern, text, flags=re.IGNORECASE)
    return len(matches)

def calculate_average_word_length(text):
    words = word_tokenize(text)
    total_words = len(words)
    total_characters = sum(len(word) for word in words)
    return total_characters / total_words if total_words > 0 else 0

# Iterating through each file in the cleaned text directory
for filename in os.listdir(input_dir):
    if filename.endswith('.txt'):
        file_path = os.path.join(input_dir, filename)
        url_id = filename.split('.')[0] 
        
        with open(file_path, 'r', encoding='utf-8') as file:
            cleaned_text = file.read()
            
            positive_score, negative_score = calculate_scores(cleaned_text, positive_words, negative_words)
            polarity = calculate_polarity(positive_score, negative_score)
            total_word_count = count_words_without_punctuation(cleaned_text)
            subjectivity = calculate_subjectivity(positive_score, negative_score, total_word_count)
            average_sentence_length = calculate_average_sentence_length(cleaned_text)
            percentage_complex_words = calculate_percentage_complex_words(cleaned_text)
            fog_index = calculate_gunning_fog_index(average_sentence_length, percentage_complex_words)
            total_syllables = count_syllables_in_text(cleaned_text)
            total_complex_words = count_complex_words(cleaned_text)
            total_personal_pronouns = count_personal_pronouns(cleaned_text)
            average_word_length = calculate_average_word_length(cleaned_text)

            # Append values to the respective lists
            positive_scores.append(positive_score)
            negative_scores.append(negative_score)
            polarity_scores.append(polarity)
            subjectivity_scores.append(subjectivity)
            average_sentence_lengths.append(average_sentence_length)
            percentage_complex_words_list.append(percentage_complex_words)
            fog_indices.append(fog_index)
            average_words_per_sentence_list.append(average_sentence_length)  # This should be updated if a different calculation is used
            complex_word_counts.append(total_complex_words)
            word_counts.append(total_word_count)
            syllables_per_word_list.append(total_syllables)
            personal_pronouns_counts.append(total_personal_pronouns)
            average_word_lengths.append(average_word_length)

# Adding new columns to the DataFrame
df['Positive Score'] = positive_scores
df['Negative Score'] = negative_scores
df['Polarity Score'] = polarity_scores
df['Subjectivity Score'] = subjectivity_scores
df['Average Sentence Length'] = average_sentence_lengths
df['Percentage of Complex Words'] = percentage_complex_words_list
df['Gunning Fog Index'] = fog_indices
df['Average Number of Words Per Sentence'] = average_words_per_sentence_list
df['Complex Word Count'] = complex_word_counts
df['Word Count'] = word_counts
df['Syllable Per Word'] = syllables_per_word_list
df['Personal Pronouns'] = personal_pronouns_counts
df['Average Word Length'] = average_word_lengths

# Updated DataFrame to the output Excel file
df.to_excel(output_excel_path, index=False)

print(f"Metrics have been successfully calculated and added to the Excel file: {output_excel_path}")
print(df)
```
